```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('C:\\Users\\hp\\Downloads\\download (1).jpeg', cv2.IMREAD_COLOR)
if img is None:
    print("Could not open or find the image")
else:
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imshow('Display window', img_gray)
    cv2.waitKey(0)
    plt.imshow(img_gray)
    plt.show()
```

```python
inverted_image = 255 - img_gray
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
plt.imshow(inverted_image)
plt.show()
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
plt.imshow(blurred)
plt.show()
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(img_gray, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
plt.imshow(inverted_blurred)
plt.show()
```
